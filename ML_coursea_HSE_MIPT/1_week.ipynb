КАК ЗАДАЮТСЯ ОБЪЕКТЫ:

-- признаковое описание  - функции как способы измерения над объектами, ну и вообще, признаки - это совокупность функций, которые возвращают некие значения, присущие этим объектам - признаки

В свою очередь, признаки делятся по типам:

- бинарные признаки - да или нет
- номинальный признак - признак, прнинимающий конечное множество значений
- порядковый признак - на множестве значений есть отношение порядка, то есть у нас возникает некий порядок в множестве значений, например - воинский звания или град студентов
- количественный признак - числовое измерение объекта

В описании одного объекта могут присутсвовать несколько типов признаков

задачи классификации: 
- нужно принять одно из двух решений
- классификация на некое количество непересекающихся классов
- классификация на некое количество пересекающихся классов

задачи восстановления регрессии:
- ответы - дествительные числа
- задачи прогнозирования (экономика, техника)

задачи ранжирования:
- множество допустимых ответов - ранжирования поисковой выдачи


каким образом выбирается предсказательная модель:

подбираем функцию или семейство функций, которые уже придуманы, которые параметрически дают нам хорошо апроксимирующийся (максимально приближенный) результат

- линейная модель регрессии - взвешенная сумма всех признаков.












```python
import numpy as np # импорт библиотеки под псевдонимом np

rand_matrix = np.random.normal(loc = 1, scale = 10, size = (1000, 50))
""" loc - среднее значение нормального распределения
    scale - стандартное отклоенение нормального распределения
    size - размер матрицы
"""
#print(rand_matrix) # вывод матрицы, не очень уж интересно 

meaning = np.mean(rand_matrix, axis = 0) # вычисляем среднее значение столбца матрицы

standart = np.std(rand_matrix, axis = 0) # вычисляем стандратное отклонение матрицы по столбцам

new_matrix = ((rand_matrix - meaning) / standart) # нормировка матрицы

import matplotlib.pyplot as plt # вызовем графикостроилку

y = plt.plot(rand_matrix)
#x = plt.plot(new_matrix)
```

```python
Z = np.array([[4, 5, 0], 
             [1, 9, 3],              
             [5, 1, 1],
             [3, 3, 3], 
             [9, 9, 9], 
             [4, 7, 1]])

sum_strings = np.sum(Z, axis = 1) # сумма элементов в строке

print (np.nonzero(sum_strings < 10)) # вывести номера элементов массива, которые больше 10
```

```python
"""
объединение матриц
"""

A_matrix = np.eye(3) # создание единичной матрицы (в которой единички по-диагонали) в скобках указывается размерность матрицы
B_matrix = np.eye(3)
stacked_matrix = np.vstack((A_matrix, B_matrix)) # vstack - вертикальная стыковка матрицы, hstack - горизонтальная стыковка матрицы
print(stacked_matrix)
```
