```python
# с помощью это команды можно увидеть разлет признаков, есть ли пропуски в данных, как это видно из каунта колонки возраста
data_from_inet.describe()
```

```python
# уникальные значения  в стобце:
set(data_from_inet["Sex"])
```

```python
# количество уникальных значений в столбце, интересно, что там такое unknown:
data_from_inet["Sex"].value_counts()
```

```python
# перекодирование стобца с помощью функции map:
sex = data_from_inet["Sex"]
sex.map({"male": 1, "female": -1, "unknown": 0}).head(5)
```

```python
# применение фунции apply в связзке с lambda функцией:
data_from_inet['NewAge'] = data_from_inet['Age'].apply(lambda x: x+100)
```

```python
data_from_inet  # новый столбик добавляется в конец датафрейма
```

```python
data_from_inet.groupby('Sex')  # группировка по каким-либо параметрам
```

```python
# группировка по среднему значения по каждому полу:
data_from_inet.groupby("Sex")["Age"].apply(lambda x: x.mean())
```

```python
# отделим фамилию в столбце с ФИО:
data_from_inet["Family"] = data_from_inet["Name"].apply(
    lambda s: s.split(";")[0])  # отделяем только первое слово
```

```python
data_from_inet.groupby("Family")["PassengerId"].count().head(
    10)  # сколько человек было с одинаковой фаимлией
```
