## группировка данных

df.groupby(by=grouping_columns)[columns_to_show].function()  
К датафрейму применяется метод groupby, который разделяет данные по grouping_columns

```python
# зададим переменную, включающую в себя выборку столбцов, которые нужно вывести:
columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[columns_to_show].describe(percentiles=[0.25]) # группировка данных в зависимости от признака churn
# процентиль - значение, ниже которого могут быть обнаружен некоторый процент наблюдений выборки
# 20 процентный процентиль покажет, сколько объектов находится ниже порога в 50 процентов.
```

```python
# то же самое можно сделать с помощью метода agg
df.groupby(['Churn'])[columns_to_show].agg([np.mean, np.max])
```

## преобразование датафреймов

```python
# создадим объект тотал колс, который будет включать в себя
# те стобцы, которые мы хотим вставить в общий датафрейм
total_calls = (df['Total day calls']+
              df['Total eve calls']+
              df['Total night calls']
             )
```

```python
df.insert(loc = len(df.columns), column='Total calls', 
          value=total_calls)
# вставка в таблицу нового столбца при помощи метода
# insert. Для вставки в конец в loc мы указали 
# len(df,columns) для подсчета количества стобцов
```

```python
# добавить столбец можно без создания новых датафреймов
df['Total charge'] = (df['Total day charge']+
                      df['Total eve charge']+
                      df['Total night charge'])
```
