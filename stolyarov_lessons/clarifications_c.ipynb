исполняемый файл на си устроен так, что после загрузки его в оперативную память управление получит код, скомпилированный из функции main
фигурные скобки определяют составной оператор

    обязательные пояснения для функции printf:
- она может выводить не только строки, но что угодно
- читает функция слева-направо
- если она встречает знак процента, то считывает символы после него, которые и будут означать, что будет выведено:
* %d - выведено целое число в десятичной системе
* %x - выведено целое число в шестнадцатеричной системе
* %05d - выведено целое число будет содержать не менее пяти знаков, а если в числе знаков меньше, то его добьют нулями до указанного значения
- %f - для печати double and float %.5f - явное указание количества знаков после запятой





```c
#include <stdio.h>

int main()
{
    int x = 17;
    float y = 0.12345;
    double p,q,r;
    int n;
        
    
    printf("x = %d and multiply of x = %d\n", x, x*x);
    printf ("%d in hex = %03x\n\n", x, x);
    printf("float output y is %.3f\n", y);
    
    printf("Please, enter 3 numbers!\n\a");
    scanf("%lf,%lf,%lf", &p, &q, &r);
    if (n!=3)
    {
        printf("Wrong input data!\n");
        return 1;
    }
}
```

        пояснения для функции scanf:
- похожа по синтаксису на функцию printf
- переменная - некая область оперативной памяти, и к ней не обязательно обращаться через ее имя, у нее также есть и адрес
- операция взятия адреса обозначается в си символом &, например, значением выражения &x будет адрес переменной x

- программа на языке си состоит из единиц трансляций - модулей (файлов), каждый из которых обрабатывается компилятором отдельно от других, то есть комплиятор не знает о том, что содержится в соседних модулях. Результатом трансляции модуля становится файл с объектным кодом, набор таких файлов связывается в готовую программу с помощью системного редактора связей или линкера.
- отдельно взятый файл, написанный на языке си состояит из глобальных объявлений и глобальных описаний. Разница между объявлением и описанием состояит в том, что объявление лишь сообщает компилятору, что объект, соответствующий данному имени где-то существует, возможно, что даже не в этом модуле. в то время как описание объекта дает полную информацию о модуле и предписывает компилятору создать объект здесь и сейчас.
