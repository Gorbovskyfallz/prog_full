```python
def is_simple_number(x):
    """определяет, является ли число простым, если простое, то возвращает Ъ, а иначе НЕЪ - 
    x - целое положительно число
    многострочная документ-строка"""
    divisor = 2
    while divisor < x:
        if x%divisor == 0:
            return False
        divisor+=1
    return True

help(is_simple_number)
is_simple_number(10)
    
    
```

```python
def factorize_number(x):
    """
        раскладывает число на множители
        печатает их на экран
        число - целое и положительное
    """
    divisor = 2
    while x > 1:
        if x%divisor == 0:
            print(divisor)
            x //= divisor # разделить нацело
        else:
            divisor +=1
            
factorize_number(999)
```

```python
#программа для отрисовки человечка на земле и фоне неба

#импорт библиотеки примитивной графики под синонимом gr для сокращения и удобства в использовании
import graphics as gr

win_universe = gr.GraphWin("Shiny universe", 400, 400) 


def earth_and_sky():
    """
    функция рисует небо и землю, состоящую из двух примитивов - прямоугольников, нижний из которых представляет землю
    и выкрашен в коричневый, верхний представляет небо и выкрашен в голубой цвет
    
    """
    #рисуем нижний прямоугольник земли:
    earth = gr.Rectangle(gr.Point(0, 400), gr.Point(400, 200))
    # зададим цвет земле:
    earth.setFill('brown3')
    # для земли зададим пожирнее линии:
    earth.setWidth(5)
    
    #рисуем верхний прямоугольник неба:
    sky = gr.Rectangle(gr.Point(0,0), gr.Point(400, 200))
    #зададим цвет небу:
    sky.setFill('cyan')
    
    
    #выводим результаты в окно:
    sky.draw(win_universe)
    earth.draw(win_universe)


def man_head():
    """
    функция рисует голову человечка - сама голова в виде круга, две брови, два глаза и рот
    
    """
    # рисуем голову, делаем ее белой и выводим в окно
    head = gr.Circle(gr.Point(200, 100), 40)
    head.setFill('white')
    head.draw(win_universe)
    
    #рисуем левый и правый глаз, задаем им цвет и выводим в окно:
    left_eye = gr.Circle(gr.Point(185, 93), 8)
    left_eye.setFill('')
    left_eye.setWidth(3)
    left_eye.draw(win_universe)
    
    right_eye = gr.Circle(gr.Point(215, 93), 8)
    right_eye.setFill('')
    right_eye.setWidth(3)
    right_eye.draw(win_universe)
    
    #рисуем рот и брови:
    
    mouth = gr.Line(gr.Point(187, 117), gr.Point(213,117))
    mouth.setWidth(3)
    mouth.draw(win_universe)
    
    left_brow = gr.Line(gr.Point(180, 77), gr.Point(190, 77))
    left_brow.setWidth(3)
    left_brow.draw(win_universe)
    
    right_brow = gr.Line(gr.Point(210, 77), gr.Point(220, 77))
    right_brow.setWidth(3)
    right_brow.draw(win_universe)

def body():
      
    """
    функция рисует тело человечка - овальное туловище, две руки и две ноги
    
    """
    #наприсуем тело
    body = gr.Oval(gr.Point(150,140), gr.Point(245,300))
    body.setFill('white')
    body.setWidth(3)
    body.draw(win_universe)
    
    #нарисуем пару ног:
    left_leg = gr.Line(gr.Point(165,280), gr.Point(165,330))
    left_leg.setWidth(3)
    left_leg.draw(win_universe)
    
    right_leg = gr.Line(gr.Point(227,280), gr.Point(227, 330))
    right_leg.setWidth(3)
    right_leg.draw(win_universe)
    
    #нарисуем пару рук:
    left_arm = gr.Line(gr.Point(149, 220), gr.Point(180, 220))
    left_arm.setWidth(3)
    left_arm.draw(win_universe)
    
    right_arm = gr.Line(gr.Point(247, 220), gr.Point(278, 220))
    right_arm.setWidth(3)
    right_arm.draw(win_universe)
       
    def painting_all():
    """
    функция для отрисовки всего и вся
    
    """
    earth_and_sky() #рисуем небо и землю
    man_head() # рисуем голову
    body() # рисуем тело
    
    # окно ожидает, пока по нему не ткнут мышкой и закрывается:
    win_universe.getMouse()
    win_universe.close()
    
painting_all() # функция рисования вообще всего, итоговая
```
